# Visualização dos Dados

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['idade'], bins=20, kde=True)
plt.title('Distribuição da Idade')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.show()
```

```python
# Importando as bibliotecas necessárias
import matplotlib.pyplot as plt

# Plotando o gráfico de barras
df['statusTrabalhista'].value_counts().plot(kind='bar', figsize=(10,5))

# Adicionando título e rótulos aos eixos
plt.title("Número de indivíduos por status de emprego")
plt.ylabel('Número de indivíduos')
plt.xlabel('Status de Emprego')

# Exibindo o gráfico
plt.show()
```

```python
df['statusTrabalhista'].value_counts().nlargest(10).plot(kind='bar', figsize=(10,5))

plt.title("Os 10 empregos mais comuns")
plt.ylabel('Número de indivíduos')
plt.xlabel('Status de Emprego')

plt.show()
```

```python
# Visualização 2: Relação entre a idade e o salário
plt.figure(figsize=(10, 6))
sns.scatterplot(x='idade', y='salario', data=df)
plt.title('Relação entre Idade e Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.show()
```

```python
# Selecionar apenas as colunas numéricas
colunas_numericas = df.select_dtypes(include=['int64', 'float64'])

# Calculando a matriz de correlação
correlation_matrix = colunas_numericas.corr()

# Configurar o tamanho da figura
plt.figure(figsize=(10, 8))

# Criar o mapa de calor com Seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")

# Adicionar título
plt.title('Matriz de Correlação')

# Mostrar o gráfico
plt.show()
```

```python
# Visualização 3: Relação entre a idade e a pontuação de crédito
plt.figure(figsize=(10, 6))
sns.scatterplot(x='idade', y='score', data=df)
plt.title('Relação entre Idade e Pontuação de Crédito')
plt.xlabel('Idade')
plt.ylabel('Pontuação de Crédito')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['entrada'], bins=20, kde=True)
plt.title('Distribuição do Valor de Entrada')
plt.xlabel('Entrada')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['precoDoApartamento'], bins=20, kde=True)
plt.title('Distribuição do Preço do Apartamento')
plt.xlabel('Preço do Apartamento')
plt.ylabel('Frequência')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='precoDoApartamento', y='gastoMensal', data=df)
plt.title('Relação entre Preço do Apartamento e Gasto Mensal')
plt.xlabel('Preço do Apartamento')
plt.ylabel('Gasto Mensal')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='precoDoApartamento', y='taxaDeJuros', data=df)
plt.title('Relação entre Preço do Apartamento e Taxa de Juros')
plt.xlabel('Preço do Apartamento')
plt.ylabel('Taxa de Juros do Apartamento')
plt.show()
```
